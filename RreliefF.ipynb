```python
import os
import sklearn_relief as relief
import pandas as pd
import numpy as np
from sklearn import preprocessing

#https://gitlab.com/moongoal/sklearn-relief/-/blob/master/src/sklearn_relief.py

#global  variables
contHR=0
conthPa=0
contRad=0
contTemp=0

contCos=0
contSin=0
contBiCos=0
contBiSin=0
contWx=0
contWy=0
```

```python
def relief_selection(path,file_name):
    #print(file_name)
    
    event_path = path+file_name    
    event=pd.read_csv(event_path,index_col=0)

    df_y = event.pop('Value (mm)')
    y = df_y.to_numpy()

        # Load some data and put it in a numpy.array matrix
    my_input_matrix = event.to_numpy()
        #my_input_matrix = my_input_matrix[12:,:]
        #df_x = pd.DataFrame(my_input_matrix)
        #print(df_x)
        #print(df_y)
        #iii=input("my_input_matrix")


        
        
        # Load the label vector
    my_label_vector = np.ravel(y,order='C')

        # Choose the best 3 features
    r = relief.RReliefF(n_features=4)

    my_transformed_matrix = r.fit_transform(my_input_matrix, my_label_vector)
        #print("Event with 3 highest ranked features:\n",my_transformed_matrix)
        
    count_ft(file_name,event,my_transformed_matrix)
        
        
        #x = input("Enter to continue\n\n")

```

```python

def count_ft(file_name,event,my_transformed_matrix):
    
    
    global contHR
    global conthPa
    global contRad
    global contTemp

    global contCos
    global contSin
    global contBiCos
    global contBiSin
    global contWx
    global contWy

    

    names = event.columns
    #print("names inside conter ft",names,"->",len(names))
    # 'Value (%)', 'Value (hPa)', 'Value (W/m2)', 'Value (Â°C)', 'Monthly sin',
    # 'Monthly cos', 'Bi-monthly sin', 'Bi-monthly cos', 'Wind X', 'Wind Y'

    

    for i in range(0,len(names)):
        
        feature =  event[[names[i]]]        
        ft = feature.to_numpy()
        ft = np.ravel(feature)
        #ft = np.ravel(feature[12:])

        for j in range(0,4):            
            vect = my_transformed_matrix[0:,j]         
            value = np.ravel(vect - ft)
            #print("value",value)            
            
            if( (np.sum(value) == 0) and (i ==0) ):                
                contHR+=1
                #print("vector is:\n",vect)
                #print("contHR = ",contHR)
                #iii=input("saved value")


            if( (np.sum(value) == 0) and (i ==1 )):
                conthPa+=1
                #print("feature",ft)
                #print("vector is:\n",vect)
                #print("conthPa = ",conthPa)
                #iii=input("saved value")

                
            if( (np.sum(value) == 0) and (i ==2) ):
                contRad+=1
                #print("vector is:\n",vect)
                #print("contRad = ",contRad)
                #iii=input("saved value")


            if( (np.sum(value) == 0) and (i ==3) ):
                contTemp+=1
                #print("feature",ft)
                #print("vector is:\n",vect)
                #print("contTemp = ",contTemp)
                #iii=input("saved value")


            if( (np.sum(value) == 0) and (i ==4) ):
                contSin+=1 
                #print("vector is:\n",vect)
                #print("contSin = ",contSin)
                #iii=input("saved value")


            if( (np.sum(value) == 0) and (i ==5) ):
                contCos+=1     
                #print("vector is:\n",vect)
                #print("contCos = ",contCos)
                #iii=input("saved value")  

            if( (np.sum(value) == 0) and (i ==6) ):
                contBiSin+=1 
                #print("vector is:\n",vect)
                #print("contBiSin = ",contBiSin)
                #iii=input("saved value")


            if( (np.sum(value) == 0) and (i ==7) ):
                contBiCos+=1     
                #print("feature",ft)
                #print("vector is:\n",vect)
                #print("contBiCos = ",contBiCos)
                #   iii=input("saved value")         
                             
            if( (np.sum(value) == 0) and (i ==8) ):
                contWx+=1 
                #print("vector is:\n",vect)
                #print("contWx = ",contWx)
                #iii=input("saved value")

                                          
            if( (np.sum(value) == 0) and (i ==9) ):
                contWy+=1 
                #print("vector is:\n",vect)
                #print("contWy = ",contWy)
                #iii=input("saved value")
```
